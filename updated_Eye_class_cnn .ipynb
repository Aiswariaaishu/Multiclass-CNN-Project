   ##                          Entropion Ectropion Eye disease Class Prediction

```python
#Importing Libreries
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
import cv2, PIL, glob, pathlib
import numpy as np
```

```python
#Accessing folder path
combined = pathlib.Path("C:/Users/user/OneDrive/Desktop/Eye_class/Train/combined")
Ectropian = pathlib.Path("C:/Users/user/OneDrive/Desktop/Eye_class/Train/Ectropion")
Entropian = pathlib.Path("C:/Users/user/OneDrive/Desktop/Eye_class/Train/Entropion")
Healthy = pathlib.Path("C:/Users/user/OneDrive/Desktop/Eye_class/Train/Healthy")
```

```python
#Using glob, taking all jpg images in each folder

images_dict = {"combined": list(combined.glob("*.jpg")),
              "Ectropian": list(Ectropian.glob("*.jpg")),
              "Entropian":list(Entropian.glob("*.jpg")),
              "Healthy":list(Healthy.glob("*.jpg"))}
labels_dict = {
    "combined":0, "Ectropian":1, "Entropian":2, "Healthy":3
}
```

```python
#Printing length of each folder
print(len(images_dict['combined']))
print(len(images_dict['Ectropian']))
print(len(images_dict['Entropian']))
print(len(images_dict['Healthy']))
```

```python
images_dict.items()
```

```python
X, y = [], []
for label, images in images_dict.items():
    for image in images:
        image = cv2.imread(str(image))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (180, 180))
        if image is not None:
            X.append(image) 
            y.append(labels_dict[label]) 
```

```python
X
```

```python
y
```

```python
X = np.array(X)
y = np.array(y)
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.imshow(X[0])
plt.show()
```

```python
X = X/255
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for image, label, ax in zip(X_train, y_train, axes.flatten()):
    ax.imshow(image)
    ax.set_title(label)
    ax.set_axis_off()
```

```python
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import models, layers
```

```python
data_argumentation = keras.Sequential([
    keras.layers.experimental.preprocessing.RandomRotation(0.2),
    keras.layers.experimental.preprocessing.RandomContrast(0.3),
    keras.layers.experimental.preprocessing.RandomZoom(0.3),
    keras.layers.experimental.preprocessing.RandomZoom(0.7)
])
```

```python
model = keras.Sequential([
    data_argumentation,
    layers.Conv2D(64, (5, 5), padding="same",   input_shape=(180, 180, 3), activation="softmax"),
    layers.MaxPooling2D(),
    layers.Conv2D(32, (5, 5), padding="same", activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(16, (5, 5), padding="same", activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(8, (5, 5), padding="same", activation="relu"),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(4, activation="softmax"),
])
```

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=5)
```

```python
training_history = model.fit(X_train, y_train, epochs=90,validation_data=(X_test, y_test))
```

```python
print(training_history.history.keys())
```

```python
plt.plot(training_history.history['accuracy'])
plt.plot(training_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(training_history.history['loss'])
plt.plot(training_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
